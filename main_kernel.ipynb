```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.kernel_ridge import KernelRidge
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
from SpotifyPredictor import RidgeRegression
from kernelridge import KernelRidgeRegression
ridge_regression_model = RidgeRegression(alpha=.3)
model = Ridge(alpha=.3)
modelK = KernelRidge(alpha=.3, kernel='rbf', gamma=1)
model_kernel_costum = KernelRidgeRegression(alpha=.3, gamma=1)
min_max_scaler = MinMaxScaler()
df = pd.read_csv('C:/Users/pietr_3fbsul4/Desktop/MLproject/SpotifyDataSet.csv').drop('Unnamed: 0', axis=1)
df = df.iloc[0:200]
ndf = df.select_dtypes(include=['int64', 'float64'])
encoder = OneHotEncoder(sparse=False, drop='first')
time_signature_encoded = encoder.fit_transform(ndf[['time_signature']])
time_signature_encoded_DF = pd.DataFrame(time_signature_encoded, columns=encoder.get_feature_names_out(['time_signature']))
ndf = pd.concat([ndf, time_signature_encoded_DF], axis=1).drop('time_signature', axis=1)
X = ndf.drop("popularity", axis=1)
y = ndf["popularity"]
kf = KFold(n_splits=5, shuffle=True)
scores_gr_custom = cross_val_score(model_kernel_costum, X, y, cv=kf, scoring= 'r2') #scikit learn regression
scores_sklearn = cross_val_score(modelK, X, y, cv=kf, scoring= 'r2')
mean_score_sk = scores_sklearn.mean()
mean_score_gr_custom = scores_gr_custom.mean()
print("scores_sklearn:", mean_score_sk)
print("mean_score_gr_custom: ",mean_score_gr_custom)


```
